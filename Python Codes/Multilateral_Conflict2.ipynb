```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
import numpy as np
R0 = 1000000
B0 = 800000
G0 = 700000
```

```python
 def model(y ,t):
  B=y[0]
  R=y[1]
  G=y[2]
  Br=0.75; Bg=0.4; Db=0.35; Dg=0.45; Yb=0.25; Yr=0.55;   
  f0 = - (Br*B)/(Br*B+Yr*G)*Db*R - (Bg*B)/(Bg*B+Dg*R)*Yb*G 
  f1 = - (Db*R)/(Db*R+Yb*G)*Br*B - (Dg*R)/(Dg*R+Bg*B)*Yr*G 
  f2 = - (Yb*G)/(Yb*G+Db*R)*Bg*B - (Yr*G)/(Yr*G+Br*B)*Dg*R 
  if B <= 0 and f0 <= 0:
    B = 0
    f0 = 0
  if R <= 0 and f1 <= 0:
    R = 0
    f1 = 0
  if G <= 0 and f2 <= 0:
    G = 0
    f2 = 0
  return [f0, f1, f2]
```

```python
y0 = [B0, R0, G0]

t = np.linspace(0, 7., 1000)

soln = odeint(model, y0, t)
B = soln[:, 0]
R = soln[:, 1]
G = soln[:, 2]

plt.figure()
plt.figure(figsize =(10,5))
plt.plot(t, B, label = "B", color = "blue")
plt.plot(t, R, label = "R", color = "red")
plt.plot(t, G, label = "G", color = "green")
plt.legend(loc = 0)
```
