```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
import numpy as np
B0 = 1000000
R0 = 1000000
```

```python
def model(y ,t):
  B=y[0]
  R=y[1]
  l1=0.3; l2=0.4
  f0 = -l1*t*R
  f1 = -((l2*t*B)/(l2+t))
  if B <= 0 and f0 <= 0:
    B = 0
    f0 = 0
  if R <= 0 and f1 <= 0:
    R = 0
    f1 = 0
  return [f0, f1]
```

```python
y0 = [B0, R0]
t = np.linspace(0, 5., 1000)

soln = odeint(model, y0, t)
B = soln[:, 0]
R = soln[:, 1]

plt.figure()
plt.figure(figsize =(10,5))
plt.plot(t, B, label = "B", color = "blue")
plt.plot(t, R, label = "R", color = "red")
plt.legend(loc = 0)
```
